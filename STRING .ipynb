```python
def reverse_string(s):
  return s[::-1]

original_string = "hello"
reversed_str = reverse_string(original_string)
print(f"Original string: {original_string}")
print(f"Reversed string: {reversed_str}")
```

```python
from collections import Counter
my_list = [1, 2, 2, 3, 1, 4, 2, 5, 3, 1, 1]

# Calculate frequency of each element in the list
list_frequency = Counter(my_list)

print("Frequency of each element in the list:")
for element, count in list_frequency.items():
    print(f"{element}: {count}")
```
